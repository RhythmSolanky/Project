```python
## TOP 10 ARTISTS ON SPOTIFY
top_ten_artists = df.groupby("artist").count().sort_values(by="song_title", ascending=False)["song_title"][:10]
top_ten_artists
```

```python
## TOP 10 SONGS ON SPOTIFY
top_ten_songs = df.groupby("song_title").count().sort_values(by="valence", ascending=False)["valence"][:10]
top_ten_songs
```

```python
top_ten_artists.plot.bar()
plt.xlabel("Artists")
plt.ylabel("No. of Songs")
plt.title("TOP 10 ARTISTS ON SPOTIFY", fontsize=16, fontweight="bold")
plt.show()
```

### TOP 10 LOUDEST TRACKS ON SPOTIFY


```python
top_ten_loudest_tracks = df[["loudness", "song_title"]].sort_values(by="loudness", ascending=True)[:10]
top_ten_loudest_tracks
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="loudness", y="song_title", data=top_ten_loudest_tracks)
plt.xlabel("Loudness")
plt.ylabel("Song titles")
plt.title("TOP 10 LOUDEST SONGS", fontsize = 16, fontweight= "bold")
plt.show()
```

### TOP 5 MOST DANCEABLE SONGS ON SPOTIFY

```python
top_five_artists_danceable_songs = df[["danceability", "song_title", "artist"]].sort_values(by="danceability", 
                                                                                            ascending=False)[:5]
top_five_artists_danceable_songs
```

```python
plt.figure(figsize=(10, 2))
sns.barplot(x="danceability", y="artist", data=top_five_artists_danceable_songs)
plt.xlabel("Danceability")
plt.ylabel("Artists")
plt.title("Artist With The Most Danceable Song")
plt.show()
```

```python
top_ten_instrumental_tracks = df[["instrumentalness", "song_title", "artist"]].sort_values(by="instrumentalness", 
                                                                                           ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(10, 5))
plt.pie(x="instrumentalness", data=top_ten_instrumental_tracks, autopct='%1.2f%%', labels=top_ten_instrumental_tracks.song_title)
plt.show()
```

```python
interest_feature_cols = ["tempo", "loudness", "acousticness", "danceability", "duration_ms", 
                         "energy", "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
for feature_col in interest_feature_cols:
    pos_data = df[df["target"] == 1][feature_col]
    neg_data = df[df["target"] == 0][feature_col]
    
    plt.figure(figsize=(12, 7))
    
    
    sns.distplot(pos_data, bins=30, label="Positive", color="blue")
    sns.distplot(neg_data, bins=30, label="Negative", color="red")
    
    
    plt.legend(loc="upper right")
    plt.title(f"Positive And Negative Histogram Plot For {feature_col}")
    plt.show()
```
